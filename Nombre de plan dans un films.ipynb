Essaie d'ecriture de code pour la détection des plans de coupe d'un film.

```python
import numpy as np
import matplotlib.pyplot as plt

filename= "cas.npy"

r=np.load(filename)
x=np.arange(r.size)
print (r,x)


```

```python
plt.plot(x,r)
plt.show()
```

Ce graphique représente la valeut de l'autocorelation pour chaque numéro de frame, on a une distribution irrégulière.

On va en tracer l'histogramme

```python
fig, ax = plt.subplots(figsize=(16, 5))
hist=plt.hist(r[~np.isnan(r)],100)
ax.set_yscale("log", nonposy='clip')
```

Pour la suite on va s'interesser au valeur superieur à $r >0,6$

## Histogramme cumulatif

```python

fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(-r[~np.isnan(r)],200,histtype='step',
        cumulative=True, range= (-1.5,-0.6) )
#




plt.show()
```

Un film a de lorde de 200 plans on va donc prendre $R = 0,7$

(0,7 pour le pére noel est une ordure et 1,2 pour casino qui fait trois heures très cut)
